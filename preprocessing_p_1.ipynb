```python



# If the nutrition column contains strings like '200,10g,5g,300mg,5g,2g,30g'
df_merged['nutrition'] = df_merged['nutrition'].apply(lambda x: x.split(','))

# Now proceed with extracting and splitting the 'nutrition' column
nutrition_data = df_merged['nutrition'].tolist()

# List of column names
columns = ['calories', 'total fat', 'sugar', 'sodium', 'protein', 'saturated fat', 'carbohydrates (PDV)']

# Convert the 'nutrition' column into separate columns with the specified headers
df_merged[columns] = pd.DataFrame(nutrition_data, columns=columns)

# Drop the original 'nutrition' column
df_merged = df_merged.drop('nutrition', axis=1)

# Display the updated DataFrame
df_merged

```

```python
df_merged['calories'] = df_merged['calories'].astype(str).str.lstrip('[')
df_merged['carbohydrates (PDV)'] = df_merged['carbohydrates (PDV)'].astype(str).str.rstrip(']')
```

```python
# Example: Filter to include only users with more than 10 ratings
user_counts = df_RAW_interactions['user_id'].value_counts()
filtered_users = user_counts[user_counts > 10].index
df_filtered = df_RAW_interactions[df_RAW_interactions['user_id'].isin(filtered_users)]
```

```python
# Downcast numerical columns to more memory-efficient types
df_RAW_interactions['user_id'] = pd.to_numeric(df_RAW_interactions['user_id'], downcast='integer')
df_RAW_interactions['recipe_id'] = pd.to_numeric(df_RAW_interactions['recipe_id'], downcast='integer')
df_RAW_interactions['rating'] = pd.to_numeric(df_RAW_interactions['rating'], downcast='float')
```

```python

# Normalize numeric features
scaler = MinMaxScaler()
nutrition_columns = ['calories', 'total fat', 'sugar', 'sodium', 'protein', 'saturated fat', 'carbohydrates (PDV)']
df_merged[nutrition_columns] = scaler.fit_transform(df_merged[nutrition_columns])

# Map user and recipe IDs to continuous indices
user_id_map = {id_: idx for idx, id_ in enumerate(df_RAW_interactions['user_id'].unique())}
recipe_id_map = {id_: idx for idx, id_ in enumerate(df_RAW_interactions['recipe_id'].unique())}

# Map IDs to indices
rows = np.array([user_id_map[id_] for id_ in df_RAW_interactions['user_id']])
cols = np.array([recipe_id_map[id_] for id_ in df_RAW_interactions['recipe_id']])
data = df_RAW_interactions['rating'].values

# Create sparse interaction matrix
n_users = len(user_id_map)
n_recipes = len(recipe_id_map)
interaction_matrix_sparse = csr_matrix((data, (rows, cols)), shape=(n_users, n_recipes), dtype=np.float32)

# Compute approximate nearest neighbors
from sklearn.neighbors import NearestNeighbors

# Using cosine metric
model = NearestNeighbors(metric="cosine", algorithm="brute")
model.fit(interaction_matrix_sparse)


```
